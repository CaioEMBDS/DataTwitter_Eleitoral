Criando fun√ß√£o de limpeza

```python
def limpeza_texto(tweet):
  tweet = tweet.lower() #Todas as letras min√∫sculas
  tweet = re.sub('\n','', tweet)
  tweet = re.sub('@[^\s]+','',tweet)
  tweet = re.sub('\B#\S+','',tweet)
  tweet = re.sub('https?://\S+|www\.\S+','',tweet)
  return tweet
```

```python
dfinal_candidatos['Tweet_limpo'] = dfinal_candidatos['Tweet'].apply(lambda i: limpeza_texto(i)) #.apply() aplica de linha a linha uma fun√ß√£o
#                                                                   fun√ß√£o lambda √©uma forma pythonica
#                                                                   de fazer fun√ß√µes em python
```
