```python
import pandas as pd

#Express√µes Regulares
import re

#Para Visualiza√ß√£o
import matplotlib.pyplot as plt
import seaborn as sns

#Para an√°lise de sentimentos
import nltk

from nltk.sentiment.vader import SentimentIntensityAnalyzer as SIA
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

##DataFrame

```python
df_sentimento = pd.read_csv('/content/Etapa 3.csv')
```

```python
df_sentimento
```

##Aplicando Fun√ß√£o de An√°lise de Sentimentos

```python
analisador_de_texto = SentimentIntensityAnalyzer()
```

```python
df_sentimento['Sentimentos'] = df_sentimento['Tweet_limpo'].apply(lambda i: analisador_de_texto.polarity_scores(' '.join(re.findall(r'\w+', i.lower()))))
```

```python
df_sentimento.head()
```

```python
df_sentimento.Sentimentos
```

##Preciso retirar da estrutura de dicion√°rios, e pegar apenas os valores e retirar as chaves com Express√µes Regulares:
>Criar 4 colunas:
>>Uma com valor do negativo

>>uma com valor do neutro

>>Uma do valor positivo

>>uma coluna final com o resultado

```python
df_sentimento['Positivo'] = df_sentimento['Sentimentos'].apply(lambda i: i['pos'])
df_sentimento.loc[df_sentimento['Positivo'] > 0] #funcionou
```

```python
df_sentimento['Neutro'] = df_sentimento['Sentimentos'].apply(lambda i: i['neu'])
df_sentimento.loc[df_sentimento['Neutro'] > 0] 
```

```python
df_sentimento['Negativo'] = df_sentimento['Sentimentos'].apply(lambda i: i['neg'])
df_sentimento.loc[df_sentimento['Negativo'] > 0] 
```

```python
df_sentimento['Composi√ß√£o'] = df_sentimento['Sentimentos'].apply(lambda i: i['compound'	])
df_sentimento.loc[df_sentimento['Composi√ß√£o']>0]
```

##Coluna com Resultado Categ√≥rico:
>precisaremos de um loop for com append()

```python
sentimento = []
```

```python
for i in df_sentimento.Composi√ß√£o:
  if i >= 0.05:
    sentimento.append('Positivo')
  elif i <= (-0.05):
    sentimento.append('Neutro')
  else:
    sentimento.append('Negativo')
```

```python
sentimento
```

```python
df_sentimento['Resultado_Sentimento'] = sentimento
df_sentimento
```

```python
df_sentimento.loc[df_sentimento['Resultado_Sentimento'] == 'Positivo']
```
